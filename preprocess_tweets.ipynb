## Fetch the original dataset
Combine all tweets per user so that each row in the dataset contains one user an all its tweets in a document

```python
import pandas as pd
```

```python
headers = ['display_name',  'username', 'user_created_at','bio', 'location', 'text_document', 'target']
```

```python
df = pd.read_csv('andrea_martine_dataset_final.csv', names=headers)
```

```python
df
```

```python
# Combine tweets per user
aggregation = {
    "username": "first",
    "display_name": "first",
    "text_document": lambda x: " ".join(x),
    "user_created_at": "first",
    "bio": "first",
    "location": "first",
    "target": "first"
}
tweet_agg = df.groupby("username").agg(aggregation)
```

```python
tweet_agg
```

```python
# Send new dataset to csv-file
tweet_agg.to_csv(r'frikk_eirik_dataset.csv', index = False)
```
